```python
import pandas as pd
import matplotlib.pyplot as plt

# Read a CSV file into a DataFrame
data = pd.read_csv('ADAS.csv')
```

```python
# To see the column names
print(data.columns)

# To display the first 20 rows of the DataFrame
print(data.head(20))
```

```python
column_61 = data.iloc[:, 60] # Warning at column 61 -> not important
print(column_61)
```

```python
ID = data['ID'] 
print(max(ID))
```

```python
Time = data['Time']
print(max(Time))
```

```python
# Plotting y-axis as 'date' and x-axis as DataFrame index
plt.plot(data.index, data['date'], marker='.', linestyle='-', color='b')
plt.title('Different dates in original data')
plt.xlabel('data index')
plt.ylabel('date')
plt.show()
```

```python
data_one_day = data[data['date'] == '2022-03-21'] #specific date
```

```python
print(data_one_day['time of day'].unique())
```

```python
extreme_case = data_one_day[data_one_day['ID']==446] #time of day with maximum number of IDs
extreme_case['time of day'].unique()
```

```python
data_one_timeday = data_one_day[data_one_day['time of day'] == '13-59-52'].reset_index() #specific time of day
```

```python
data_one_timeday
```

```python
# Plotting y-axis as values of 'ID' and x-axis as DataFrame index
plt.plot(data_one_timeday.index, data_one_timeday['ID'], marker='.', linestyle='-', color='b')
plt.title('Different IDs in a specific date and specific time of day')
plt.xlabel('index')
plt.ylabel('ID')
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(20, 20))  # Adjust the values as needed

# Creating a scatter plot with y-axis as IDs and x-axis as Timestamps
plt.scatter(data_one_timeday['Time'], data_one_timeday['ID'], marker='.', color='b')

# Set plot title and labels
plt.title('Timestamps vs IDs in each time of day')
plt.xlabel('Timestamps')
plt.ylabel('ID')

# Show the scatter plot
plt.show()
```

**Conclusion: The Timestamps(Time) and ID's are being reset in each time of day**

```python
data_one_timeday_timestamp = data_one_timeday[data_one_timeday['Time'] == 10].reset_index() #specific Timestamp
```

```python
data_one_timeday_timestamp
```

```python
# Set the figure size
plt.figure(figsize=(10, 5))  # Adjust the values as needed

# Creating a scatter plot with y-axis IDs and x-axis as DataFrame index
plt.scatter(data_one_timeday_timestamp.index, data_one_timeday_timestamp['ID'], marker='.', color='b')

# Set plot title and labels
plt.title('Different IDs at each timestamp')
plt.xlabel('index')
plt.ylabel('ID')
plt.grid(True)
# Show the scatter plot
plt.show()
plt.tight_layout()
```

**Conclusion: at each timestamp we can have at most two instances from each ID**
